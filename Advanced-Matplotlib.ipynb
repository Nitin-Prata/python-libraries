### Advanced Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal length')
plt.ylabel('patel length')
plt.colorbar()
plt.show()
```

```python
# cmap and alpha
```

### plot size

```python
plt.figure(figsize=(15,8)) # width and height
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal length')
plt.ylabel('patel length')
plt.colorbar()
plt.show()
```

### Annotations
Graph ke upar text ko place kar sakte ho

```python
batters = pd.read_csv('batter (1).csv')
batters.sample(5)
# avg means par match kitne run banayege and strike rate means 100 ball par kitne run banayege
```

```python
batters.shape
```

```python
sample_df = batters.head(100).sample(25, random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
x = [1,2,3,5]
y = [4,6,7,8]
plt.scatter(x,y)
plt.text(1,4,'point 1')
plt.text(2,6,'point 2')
plt.text(3,7,'point 3')
plt.text(5,8,'point 4',fontdict={'size':12,'color':'brown'})
```

### Horizontal and vertical lines

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='red')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```
