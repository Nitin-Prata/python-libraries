### Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}).infer_objects(copy=False)
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal length')
plt.ylabel('patel length')
plt.colorbar()
plt.show()
```

### Annotations
Graph ke upar text ko place kar sakte ho

```python
batters = pd.read_csv('batter (1).csv')
batters.sample(5)
# avg means par match kitne run banayege and strike rate means 100 ball par kitne run banayege
```

```python
batters.shape
```

```python
sample_df = batters.head(100).sample(25, random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
x = [1,2,3,5]
y = [4,6,7,8]
plt.scatter(x,y)
plt.text(1,4,'point 1')
plt.text(2,6,'point 2')
plt.text(3,7,'point 3')
plt.text(5,8,'point 4',fontdict={'size':12,'color':'brown'})
```

### Horizontal and vertical lines

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='red')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y =  np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

b = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(b)
```

```python
z = np.sin(xx) + np.log(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

b = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(b)
```

###  Contour Plots
3d graph ko 2d mai represent karta hai

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### HeatMap
+ Graph ko grid mai convert karte hai
+ and Grid batata hai ki value kaise distributed hai us grid mai
+ use case :  2 catagorical variable ko compare karte hai side by side 

```python
df = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
df.head()
```

```python
temp_df = df[(df['ballnumber'].isin([1,2,3,4,5,6])) & (df['batsman_run']==6)]
```

```python
temp_df.head(5)
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(20,20))
plt.imshow(grid)
plt.colorbar()
```
