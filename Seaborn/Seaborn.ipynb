```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# scatter plot -> genreally plot on numerical quantities
sns.scatterplot(tips,x='total_bill',y='tip')
# axes level function
```

```python
sns.relplot(tips,x='total_bill',y='tip',kind='scatter')
# relplot -> figure level -> square shape
# relplot genreal kind of function u have to change the only kind value
# figure wale function top par hote hai unke andar axes wale fucntion hote hai
```

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex')
```

```python
sns.scatterplot(tips,x='total_bill',y='tip',style='time',hue='sex')
```

```python
sns.scatterplot(tips,x='total_bill',y='tip',style='time',hue='sex',size='size')
```

```python
# scatter using replot -> size and hue
```

```python
# line plot -> basically a scater plot hi hota hai
# so agar apko continutiy show karna hai to line plot use hota hai otherwise line plot 
# or if u have time on x axes so we prefer line plot

gap = px.data.gapminder()
gap.head(5)
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level fun
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# using relpolt
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country')
# legend graph ke bahar hota hai coz yeh ek figure level fun hai
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country')
# legend graph ke andar hota hai coz yeh ek axis level fun hai
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country',style='continent',size='continent')
```

```python
# facet plot -> isko use karke particular catagorical column ke upar multiple plot bana skate ho
# work only with figure level fun -> work with rel plot
# will not work with sactter plot fun and line plot fun
sns.relplot(data=tips , x='total_bill',y='tip',kind='scatter',hue='sex')
```

```python
sns.relplot(data=tips , x='total_bill',y='tip',kind='scatter',col='sex')
```

```python
sns.relplot(data=tips , x='total_bill',y='tip',kind='scatter',row='sex')
```

```python
sns.relplot(data=tips , x='total_bill',y='tip',kind='scatter',col='smoker' , row='sex')
```

```python
sns.relplot(data=tips , x='total_bill',y='tip',kind='scatter',col='sex' , row='day')
```

```python
sns.relplot(data=tips , x='total_bill',y='tip',kind='line',col='sex' , row='day')
```

```python
# col wrap
gap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year')
```

```python
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

### 2. Distribution Plots  -> tells distrtibution of data,how data is divided

- used for univariate analysis
- used to find out the distribution
- Range of the observation (min,max)
- Central Tendency (mean,median)
- is the data bimodal? (multiple peak)
- Are there outliers?

Plots under distribution plot

- histplot -> histogram
- kdeplot
- rugplot

```python
# figure level fun -> Displot
# axes level -> histplot -> kdeplot->rugplot
```

```python
# histplot
# plotting univariate histogram -> data mai bins create hote hai then we check ki konsa value konse bin mai aata hai
# or data ka distribution batata hai,bins mai convert karke ki data ka kitna count hai us bin mai
tips
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
# bins parameter
sns.displot(data=tips,x='total_bill',kind='hist',bins=6)
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(data=tips, x='day', kind='hist') 
```

```python
sns.displot(data=tips, x='tip', kind='hist',hue='sex') 
# hue parameter -> 2 diffrent catagories ke beach mai histogram kesa hai
```

```python
# element -> step
sns.displot(data=tips, x='tip', kind='hist',hue='sex',element='step') 
```

```python
# faceting using col and row -> not work on histplot function
# element -> step
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step') 
```

```python
# kdeplot -> kernal density estimation
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.kdeplot(data=tips,x='total_bill',hue='sex',fill=True)
```

```python
# Rugplot -> same info ko alag form mai show karta hai

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### 3. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap -> 
# plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level fun
plt.figure(figsize=(25,25))
sns.heatmap(temp_df)
```

```python
# annot
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(temp_df,annot=True,linewidths=0.5,cmap='viridis')
```
