### 2D Line Plot




+ 2d plot 2 column ke upar apply hoga,single column ke upar ni (Bivariate Analysis)
+ kis type ke column ke liye useful hai -> categorical -> numerical and numerical -> numerical
+ kis type ke data ke liye useful hai ->  Use case - Time series data

```python
# plotting a simple function
price = [48000,54000,64000,49000,98000] # numerical data
year = [2015,2017,2019,2022,2025] # catagorical data

plt.plot(year,price) # x axis-> catagorical data , y axis -> numerical data
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli (1).csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# lablels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit sharma Vs Virat kohli Career Comparision')
plt.xlabel('Season')

plt.ylabel('Runs Score')
```

```python
# Colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')

plt.title('Rohit sharma Vs Virat kohli Career Comparision')
plt.xlabel('Season')

plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashdot',linewidth=2)

plt.title('Rohit sharma Vs Virat kohli Career Comparision')
plt.xlabel('Season')

plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2,marker='D')

plt.title('Rohit sharma Vs Virat kohli Career Comparision')
plt.xlabel('Season')

plt.ylabel('Runs Score')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2,marker='D',label='Rohit')

plt.title('Rohit sharma Vs Virat kohli Career Comparision')
plt.xlabel('Season')

plt.ylabel('Runs Score')

plt.legend(loc='best') # screen par lables ko print karata hai
```

### Scatter Plot


+ Use for Bivariate Analysis
+ numerical vs numerical (both column should be numerical)
+ Use case - Finding correlation between qunatities

### Note: 2d plot ko plot krne ke liye pahle scatter plot , plot hota hai fir usse joh point aate hai unke beach mai line draw kar dete hai

2d plot is the special case of scatter plot, scatter plot is the main guy

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)
```

```python
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('batter (1).csv')
df = df.head(50)
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='+')
plt.title('Avg and SR analysis of top 50 Batsman')
plt.xlabel('Avegrage')
plt.ylabel('SR')
```

```python
# marker
```

```python
# size 
# slower technique
tips = sns.load_dataset('tips') # data set about restraunt
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
# 3d main info represernt kar raha hai,total bill, total tip and us particular customer ke sath or kitne customer aaye
```

```python
# scatterplot using plt.plot
# faster technique
plt.plot(tips['total_bill'],tips['tip'],'o')
```

### Bar Chart 


+ Bivariate Analysis and Univariate Analysis
+ Numerical(Aggregarte function) vs Categorical
+ Use case - Aggregate analysis of groups

```python
# simple bar chart
child = [10,20,30,15,35]
color = ['red','blue','green','yellow','pink']

plt.bar(color,child,color='black')
```

```python
# bar chart using data
```

```python
# horizontal bar chart
# we have two type of bar chart 1. vertical(no of categories kam ho toh) and 2.horizontal(no of catagoreis jayda ho toh)
plt.barh(color,child,color='black')
```

```python
# color and label
df = pd.read_csv('batsman_season_record (1).csv')
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
# multiple Bar Charts
```

```python
# a problem
# simple bar chart
child = [10,20,30,15,35]
color = ['red red red red','blue blue blue blue','green','yellow','pink']

plt.bar(color,child,color='black')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# stacked bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2016']+df['2015']),label='2017')

plt.legend()
plt.show()
```

### pie chart


+ Univariate/Bivariate Analysis
+ Categorical vs numerical
+ Use case - To find contibution on a standard scale

### note : pie chart categries ke upar plot hota hai and yeh info deta hai ki konse categorie ka 100% mai kitna contribution hai

```python
# simple data
data = [23,45,67,12,90]
sub = ['math','eng','bio','hindi','tamil']
plt.pie(data,labels=sub)
plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175 (1).csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode=[0.1, 0, 0, 0.1, 0, 0],shadow=True)
plt.show()
```
